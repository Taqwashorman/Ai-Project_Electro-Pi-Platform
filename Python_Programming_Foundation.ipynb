```python

class DataPrepKit:

      def __init__(self, data_path):
        self.path = self.data_read(data_path)
        self.data = None

    # Function to read data from different file formats such as CSV, Excel, and JSON.
      def data_read(self, data_path):
        path = None  # Initialize datapath with None
        if data_path.endswith('.csv'):  # Corrected the file extension typo 'cvs' to 'csv'
            path = pd.read_csv(data_path)
        elif data_path.endswith('.json'):
            path = pd.read_json(data_path)  # Corrected the typo 'read_Json' to 'read_json'
        elif data_path.endswith('.xlsx'):
            path = pd.read_excel(data_path)
        else:
            raise ValueError("Unsupported file format. Only CSV, Excel, and JSON files are supported.")
        return path




#Function to summaries of the data.
      def data_summary(self):
        self.data = self.path.describe()
        print("Summary Statistics:",  self.data)
        return self.data




#Function to handle missing values.
      def handle_missing_values(self,strategy):
        if strategy == 'mean':
            self.path.fillna(self.path.mean(), inplace=True)
        elif strategy == 'median':
            self.path.fillna(self.path.median(), inplace=True)
        elif strategy == 'mode':
            self.path.fillna(self.path.mode().iloc[0], inplace=True)
        else:
            raise ValueError("Invalid strategy. Please choose from  'mean', or 'median' or 'mode'.")
            return path



#Function to encode the data.
      def encode_categorical_data(self):
        flat_data = np.ravel([str(item) for item in self.path])
        encoder = LabelEncoder()
        encoded_data = encoder.fit_transform(flat_data)
        print("Encoded data type:", encoded_data.dtype)
        print("\nEncoded data types value counts:")
        print(np.unique(encoded_data, return_counts=True))



# Display data info
      def show_data_info(self):
        print(self.path.isnull().sum())
```

```python
#/content/Book1.csv
data_path = input("Enter the path to your data file: ")
prep = DataPrepKit(data_path)

prep.data_summary()

strategy=input("plese choose the strategy for handle misssing value as (mean), (medin), (mode)")
prep.handle_missing_values(strategy)
prep.show_data_info()

print("\n Categorical data encoding:")
prep.encode_categorical_data()
```
